
# 一 轮廓初步认知  

图像轮廓用来描述图像中连续的点，它们有同样的颜色和灰度级。为了更精确地进行检测，在查找轮廓前需要先将图像做二值化处理或者使用canny边沿检测。在OpenCV中轮廓检测只查找白色目标，黑色背景会被忽略。


## 1.cvtColor灰度化图片  
```
cvtColor(图片,转换格式)  
```

````python
import warnings
warnings.filterwarnings("ignore") 
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (9,9) 
plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)  

#解决中文显示问题
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
````

````python
import numpy as np
from PIL import Image
import cv2 
img=cv2.imread('./images/class.png')
img_bg=np.zeros(img.shape, np.uint8)
imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
Image.fromarray(imgray)
````

## 2.threshold二值化  
```
 cv2.threshold(图像,开始阈值,结束阈值,方法选择)  
```

````python
ret, thresh = cv2.threshold(imgray, 127, 255, 0)
````

## 3.findContours 轮廓提取  
```
findContours(二值化图片,轮廓的检索模式,轮廓的近似办法)  

```

````python
# 节点获取
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE) 
hierarchy.shape,len(contours[0])
````

## 4.drawContours绘制轮廓  
```
cv2.drawContours(图像,轮廓,指定轮廓(-1全部),颜色,线宽（-1填充）)  
```

````python
# 所有轮廓
img_1=cv2.drawContours(img_bg.copy(), contours, -1,(255,0,0),2)
Image.fromarray(img_1)
````

## 5.单一轮廓绘画 

````python
# 单一轮廓
img_2=cv2.drawContours(img_bg.copy(), contours,4, (255,0,0),1) 
Image.fromarray(img_2)
````

# 二 轮廓检测模式  
| 轮廓检索模式 | 作用  |  
| :--- | :--- |  
| RETR_EXTERNAL | 只检测最外围轮廓，包含在外围轮廓内的内围轮廓被忽略； |  
| RETR_LIST | 检测所有的轮廓，包括内围、外围轮廓，不建立等级关系 |  
| RETR_CCOMP | 检测所有的轮廓，只建立两个等级关系，外围为顶层，若外围内的内围轮廓还包含了其他的轮廓信息，则内围内的所有轮廓均归属于顶层； |  
| RETR_TREE | 检测所有轮廓，所有轮廓建立一个等级树结构。外层轮廓包含内层轮廓，内层轮廓还可以继续包含内嵌轮廓。 |  


````python
contours, hierarchy = cv2.findContours(thresh,  cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
img_1=cv2.drawContours(img_bg.copy(), contours, -1, (255,0,0), 2)
Image.fromarray(img_1)
````

````python
contours, hierarchy = cv2.findContours(thresh,  cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
img_1=cv2.drawContours(img_bg.copy(), contours, 1, (255,0,0), 2)
Image.fromarray(img_1)
````

# 三 轮廓近似方法  
| 轮廓近似方法 | 作用  |  
| :--- | :--- |  
| CHAIN\_APPROX\_NONE | 保存物体边界上所有连续的轮廓点到contours向量内； |  
| CHAIN\_APPROX\_SIMPLE | 仅保存轮廓的拐点信息，把所有轮廓拐点处的点保存入contours向量内，拐点与拐点之间直线段上的信息点不予保留； |  
| CHAIN\_APPROX\_TC89_L1 | 使用teh-Chinl chain 近似算法; |  
| CHAIN\_APPROX\_TC89_KCOS | 使用teh-Chinl chain 近似算法。 |

````python
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
print('hierarchy.shape:',hierarchy.shape)
print('len(contours):',len(contours))
print('contoursy[0].shape:',contours[0].shape)
img_1=cv2.drawContours(img_bg.copy(), contours, -1, (255,0,0), 2)
Image.fromarray(img_1)
````

````python
import numpy as np
import cv2
im = cv2.imread('./images/cards.png')
imgray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

#需要注意的是cv2.findContours()函数接受的参数为二值图，即黑白的（不是灰度图）
# 所以读取的图像要先转成灰度的，再转成二值图
ret, thresh = cv2.threshold(src=imgray, thresh=244, maxval=255, type=cv2.THRESH_BINARY)#src, thresh, maxval, type

#轮廓提取模式 Contour_Retrieval_Mode
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
print('len(contours)', len(contours))

ct_img= cv2.drawContours(im.copy(), contours, -1, (0,255,0), 3)

titles = ['Original Image', "imgray",'thresh',"all contours"]
images = [im, imgray, thresh,ct_img] 

for i in range(4):
    plt.subplot(2, 2, i + 1), plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.subplots_adjust(wspace = 0.1, hspace =-0.2 )#调整子图间距   
plt.show()
````

# 四 轮廓特征: 边界矩形   

- 直边矩形

它是一个直的矩形，它不考虑对象的旋转。因此，边界矩形的面积不是最小的。它由函数cv.boundingRect()找到。  

- 旋转矩形 

以最小面积绘制边界矩形，因此它也考虑旋转。使用的函数是cv.minAreaRect()。

````python
img = cv2.imread('./images/lightning.png')
imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

contours, hierarchy = cv2.findContours(imgray, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours = [cnt for cnt in contours if cv2.contourArea(cnt) > 100]
print('len(contours)', len(contours))

img_contours = cv2.drawContours(img.copy(), contours, -1, (0,0,255), 3)

cnt = contours[0]
#方向是物体定向的角度
(x, y), (MA, ma), angle = cv2.fitEllipse(cnt)
print((x, y), (MA, ma), angle)

rect = cv2.minAreaRect(cnt)
box = cv2.boxPoints(rect)
box = np.int0(box)
x,y,w,h = cv2.boundingRect(cnt)
# 直边矩形               
img_cnt1= cv2.rectangle(img.copy(),(x,y),(x+w,y+h),(0,255,0),2)
# 旋转矩形
img_cnt2 = cv2.drawContours(img.copy(), [box], 0, (0, 0, 255), 2)


titles = ['Original Image', "all contours",'img_cnt1','img_cnt2']
images = [img, img_contours, img_cnt1,img_cnt2]

for i in range(4):
    plt.subplot(2, 2, i + 1), plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
````
