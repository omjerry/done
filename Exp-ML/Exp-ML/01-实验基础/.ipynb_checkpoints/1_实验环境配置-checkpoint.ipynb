## 1.2 pip切换国内源 
**pip 是 Python包管理工具**，conda 是一个开源的软件包管理系统和环境管理系统。 

本课程使用pip进行Python包管理，如需虚拟环境开发，推荐 `pip + pyenv`方式。

`python -m pip ...` 其中python是指定版本Python,

如存在多个版本Python编译器，注意Python版本，

如只有一个版本Python，可省略python -m。



pip常见用法

- 安装
`pip install requests`
- 指定版本`pip install request==版本号 `
- 卸载 `pip uninstall requests`
- 升级 `pip install --upgrade requests`
